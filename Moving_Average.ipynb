```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/skripsi/dataset/data.csv')
print(df)
df.head(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
df['prediksi'] = df.iloc[:,0].rolling(window=3).mean()
df.head(10)
```

```python
df['prediksi'] = df['prediksi'].replace(np.nan,0)
df.head()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error (df['prediksi'], df['Tanggal;total_precipitation'])
rmse=np.sqrt(mse)
print ("MSE: ",mse)
print ("RMSE: ",rmse)
```

```python
a=range(len(df))
plt.scatter(a,df['Tanggal;total_precipitation'], color = 'blue')
plt.plot(a,df['prediksi'], color = 'green')
plt.title('Curah Hujan di Brebes')
plt.xlabel('Data ke-')
plt.ylabel('Curah Hujan')
plt.show()
```
