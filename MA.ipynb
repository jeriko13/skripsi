```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
ds = pd.read_csv('/content/drive/MyDrive/skripsi/dataset/data.csv')
print(ds)

```

```python

ds['prediksi2'] = ds.iloc[:,0].rolling(window=2).mean()
ds.head(10)
```

```python

ds['prediksi3'] = ds.iloc[:,0].rolling(window=3).mean()
ds.head(10)
```

```python

ds['prediksi4'] = ds.iloc[:,0].rolling(window=4).mean()
ds.head(10)
```

```python
import numpy as np
ds=ds.replace(np.nan,0)
ds.head(10)
```

```python
from sklearn.metrics import mean_squared_error
mse1 = mean_squared_error (ds['Tanggal;total_precipitation'], ds['prediksi2'])
rmse1=np.sqrt(mse1)
print ("MSE: ",mse1)
print ("RMSE: ",rmse1)
```

```python
from sklearn.metrics import mean_squared_error
mse2 = mean_squared_error (ds['Tanggal;total_precipitation'], ds['prediksi3'])
rmse2=np.sqrt(mse2)
print ("MSE: ",mse2)
print ("RMSE: ",rmse2)
```

```python
from sklearn.metrics import mean_squared_error
mse3 = mean_squared_error (ds['Tanggal;total_precipitation'], ds['prediksi4'])
rmse3=np.sqrt(mse3)
print ("MSE: ",mse3)
print ("RMSE: ",rmse3)
```

```python
import matplotlib.pyplot as plt
a=range(len(ds))
plt.scatter(a,ds['Tanggal;total_precipitation'], color = 'green')
plt.plot(a,ds['prediksi2'], color = 'purple')
plt.title('Prediksi Curah Hujan di Brebes')
plt.xlabel('Data ke-')
plt.ylabel('Curah Hujan')
plt.show()
```
